# Acceso a ficheros

## 1.1 Ficheros

- File es un tipo de objeto predefinido en Python (built-in).
- Permite acceder a ficheros desde programas Python.
- Los ficheros son de un tipo especial:
  - Son built-in, pero no son ni números, ni secuencias, ni mappings. Tampoco responden a operadores en expresiones.
- La función `open` permite crear objetos de tipo fichero.

Formato general para abrir un fichero:

```python
afile = open(filename, mode)
```

- `mode` es opcional. Por defecto, los ficheros se abren en modo lectura.
- Los datos leídos de un fichero siempre se obtienen en formato string. Lo mismo ocurre con escritura.
- Los ficheros se deben cerrar invocando `close` (liberación de recursos).

````python
 # Lectura desde fichero usando método 'read'. Devuelve todo el contenido del fichero.
mi_fichero = open('res/res/multiple_lines.txt')
print(mi_fichero.read())
mi_fichero.close()
````

````python
# Lectura linea a linea a través del bucle 'for'.
mi_fichero = open('res/res/multiple_lines.txt')
for linea in mi_fichero:
    print(f"{linea}")
mi_fichero.close()
````

````python
# leer de un archivo a una lista.
mi_fichero = open('res/res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()
print(lineas)
````

* Resolución de paths independiente de plataforma


````python
import os
ruta = os.path.join("res","res", "multiple_lines.txt")
print(ruta)
````

````python
dataset_file_path = ["res","res","multiple_lines.txt"]
ruta = os.path.join(*dataset_file_path)
print(ruta)
````

* close autómatico con sentencia with. Esta es la forma habitual de leer de fichero en Python,
context manager

````python
with open(ruta) as mi_fichero:
    for linea in mi_fichero:
        print(linea, end= '')
````

* Se puede abrir varios ficheros en un mismo with

````python
#abrir varios ficheros en el mismo with
ruta1 = os.path.join("res","res", "one_line.txt")
ruta2 = os.path.join("res","res", "multiple_lines.txt")
with open(ruta1) as fichero1, open(ruta2) as fichero2:
    print(fichero1.readlines())
    print(fichero2.readlines())
````

### Acceso para escritura

````python
def crear_lista(tamanyo):
    lista = []
    for i in range(tamanyo):
        lista.append(str(i) + '\n')
    return lista
ruta = os.path.join("res","res", "c_dummy.txt")
with open(ruta, 'wt') as fichero:
    fichero.write('Cabecera del ciclo del for en el ejemplo\n')
    lista = crear_lista(10)
    fichero.writelines(lista)
    print("archivo creado")
````

### Buffering
* Por defecto, el texto que transfieres desde tu programa a un fichero no se guarda en disco
inmediatamente. Se almacena en un buffer.
* Acciones como cerrar un fichero o invocar el método flush fuerzan que se transfiera el contenido del buffer a disco.

````python
ruta = os.path.join("res", "FicheroParaEscritura.txt")
fichero_write = open(ruta, 'w')
fichero_write.write('foter')
fichero_read = open(ruta, "r")
print(fichero_read.readlines())
fichero_write.flush()
print(fichero_read.readlines())
fichero_write.close()
fichero_read.close()
````

### Archivos CSV
* Python permite leer datos de ficheros CSV y también escribir ficheros en este formato.
* Popular formato en ciencia de datos.

````python
# tabla_operaciones.csv contiene valores separados por comas
import csv # importar la libreria csv
ruta = os.path.join("res","res","tabla_operaciones.csv")
with open(ruta) as fichero:
    data_reader = csv.reader(fichero, delimiter=',')
    for linea in data_reader:
        print(linea[0] + ' ---- ' + linea[1])
````

````python
ruta = os.path.join("res", "res", "tabla_operaciones.csv")
f = open(ruta)
data_reader = csv.reader(f, delimiter = ',')
ruta_o = os.path.join("res","res", "4_tabla_operaciones.csv")
with open(ruta_o, 'w') as f2:
    csv_writer = csv.writer(f2, delimiter = '|')
    for line in data_reader:
        print(line)
        csv_writer.writerow(line)
f.close()
````

### Docstrings
* Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea de
comandos o durante la ejecución del programa.
* Los docstrings se almacenan en el atributo __doc__ de cada objeto.
* El valor de dicho atributo se puede mostrar por medio de la función help.


````python
help(open)
````

* No es necesario editar este atributo directamente.
* Para asociar un docstring a un objeto basta con escribir el texto (entre triples comillas) al
principio de los modulos, funciones o clases, antes del codigo ejecutable.

````python
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass
help(funcion_de_prueba)
````

*  Accediendo al atributo __doc__ sólo se obtiene el docstring.

````python
print(funcion_de_prueba.__doc__)
````

* Como se puede observar, help añade información adicional. Por ejemplo, para entidades más
grandes, help muestra el docstring dividido en secciones

````python
import sys
help(sys)
````

* Los docstrings que ocupan más de una línea suelen tener estas partes:
* Resumen en la primera línea.
* Línea en blanco.
* Descripción detallada.
* Otra línea en blanco antes del código.


````python
def funcion_de_prueba2():
    """Esta es la línea para el resumen.
    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. Observa que el resumen y la descripción
    detallada van separados por una línea en blanco. También hay
    otra línea en blanco antes de empezar el código.
    """
    pass
help(funcion_de_prueba2)
````

Algunos comentarios adicionales:

## 2.3 Ejercicios

1. Escribe una función que reciba una ruta de un fichero de texto y una cadena de caracteres a buscar y determine si la cadena aparece en el fichero.

2. Escribe una función que reciba una lista, una ruta destino y un número `n`.

La función debe crear un fichero en la ruta especificada.

El contenido del fichero serán los primeros `n` elementos de la lista.

La función debe controlar de manera apropiada los posibles valores de `n` que estén fuera de rango.

3. Escribe una función que reciba una ruta de un fichero de texto y devuelva un diccionario con la frecuencia de aparición de cada palabra.

Ejemplo: un fichero que contenga la frase:

```text
'es mejor que venga que que no venga'
```

devolverá el siguiente diccionario:

```python
{'es': 1, 'mejor': 1, 'que': 3, 'venga': 2, 'no': 1}
```

Para dividir un string en palabras puedes hacer uso del método `split`.

````python
# ejercicio 1 
def buscar_cadena(ruta_fichero, cadena):
    """
    Determina si una cadena aparece dentro de un fichero de texto.

    Args:
        ruta_fichero: Ruta del fichero de texto.
        cadena: Cadena de caracteres a buscar.

    Returns:
        True si la cadena aparece en el fichero, False en caso contrario.
    """
    try:
        with open(ruta_fichero, "r") as fichero:
            contenido = fichero.read()
            return cadena in contenido

    except FileNotFoundError:
        print("El fichero no existe.")
        return False
        
resultado = buscar_cadena("archivo.txt", "Python")

if resultado:
    print("La cadena fue encontrada.")
else:
    print("La cadena no fue encontrada.")
````

````python
# ejercicio2 
def crear_fichero(lista, ruta_destino, n):
    """
    Crea un fichero con los primeros n elementos de una lista.

    Args:
        lista: Lista de elementos.
        ruta_destino: Ruta donde se creará el fichero.
        n: Número de elementos a escribir.

    Returns:
        None
    """
    try:
        if n < 0 or n > len(lista):
            print("El valor de n está fuera de rango.")
            return

        with open(ruta_destino, "w") as fichero:
            for elemento in lista[:n]:
                fichero.write(str(elemento) + "\n")

        print("Fichero creado correctamente.")

    except Exception as error:
        print("Error al crear el fichero:", error)

datos = ["Python", "Java", "C++", "JavaScript"]

crear_fichero(datos, "lenguajes.txt", 3)
````

````python
# ejercicio 3 
def frecuencia_palabras(ruta_fichero):
    """
    Devuelve un diccionario con la frecuencia de aparición
    de cada palabra en un fichero.

    Args:
        ruta_fichero: Ruta del fichero de texto.

    Returns:
        Diccionario con palabras y frecuencia.
    """
    frecuencias = {}

    try:
        with open(ruta_fichero, "r") as fichero:
            contenido = fichero.read()

            palabras = contenido.split()

            for palabra in palabras:
                if palabra in frecuencias:
                    frecuencias[palabra] += 1
                else:
                    frecuencias[palabra] = 1

        return frecuencias

    except FileNotFoundError:
        print("El fichero no existe.")
        return {}


resultado = frecuencia_palabras("texto.txt")

print(resultado)
````
