```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)
nivel_estudios = np.random.permutation(nivel_estudios)
```

```python
print(nivel_estudios)
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np
estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1. CREACIÓN ALEATORIA DE DATOS
np.random.seed(42)
nivel_estudios_raw = (
    ['Secundaria'] * 110 +
    ['Bachillerato'] * 172 +
    ['Grado'] * 145 +
    ['Master'] * 52 +
    ['Doctorado'] * 21
)
# Crear un DataFrame con pandas
df = pd.DataFrame({
    'nivel_estudios': np.random.permutation(nivel_estudios_raw)
})
print(df)
print(f"\nTotal registros: {len(df)}")
```

```python
# FRECUENCIAS CON PANDAS
frecuencias = df['nivel_estudios'].value_counts().sort_index()
print("\nFrecuencias:")
print(frecuencias)
```

```python
#3. DIAGRAMA DE BARRAS CON PANDAS
ax = frecuencias.plot(
    kind='bar',
    color='steelblue',
    edgecolor='white',
    figsize=(8, 5)
)
ax.set_title('Diagrama de barras')
ax.set_xlabel('Nivel de estudios')
ax.set_ylabel('Frecuencia absoluta')
ax.tick_params(axis='x', rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 4. DIAGRAMA DE SECTORES (PIE) CON PANDAS
colores = ['red', 'blue', 'orange', 'green', 'cyan']
ax2 = frecuencias.plot(
    kind='pie',
    colors=colores,
    autopct='%1.2f%%',
    figsize=(7, 7),
    ylabel=''           # elimina la etiqueta "nivel_estudios" del eje y
)
ax2.set_title('Diagrama de sectores')
plt.tight_layout()
plt.show()
```
