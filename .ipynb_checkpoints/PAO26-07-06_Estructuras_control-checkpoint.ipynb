Iteración (bucles)

- Repetición de un bloque de código.
- La terminación del bucle depende del tipo de bucle.
- Dos tipos: `while` y `for`.

Bucle `while`

- Repetición de un bloque de código hasta que se deje cumplir una expresión (es decir, hasta que una condición evalue a `False`).
- Si la condición evalua a `False` desde el principio, el bloque de código nunca se ejecuta.
- Cuidado con los bucles infinitos.

Formato general:

```python
while test:  # Mientras se cumple la condición
    statements  # Instrucciones a ejecutar
```

````python
 # Ejemplo: Mostrar los primeros 3 objetos de una lista
indice = 0
numeros = [9, 4, 7, 1, 2]
while indice < 3:
    print(numeros[indice])
    indice =indice+1
````

````python
# Ejemplo: contar y mostrar los números inferiores a 10.
numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
````

````python
nombre = 'Pablo'

while nombre:           # Mientras 'nombre' no sea vacío
    print(nombre)       # Imprime el contenido de nombre
    nombre = nombre[1:] # Elimina el primer carácter
````

````python
#bucle infinito
i = 0

while i < 10:
    print(i)
    i += 1
````

Bucle `for`

- Permite recorrer los items de una sequencia o un objeto iterable.
- Funciona en strings, listas, tuplas, etc.

Formato general:

```python
for item in objeto:  # Asigna los items del objeto a la variable item en cada iteración
    statements  # Instrucciones a ejecutar
```

````python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']

for pelicula in peliculas:
    print(pelicula)
````

* También se usa para iterar un número preestablecido de veces (counted loops):

````python
for i in range(10):
    print(i)
````

 * range es útil en combinación con len porque permite acceder a los elementos de una secuencia
por posición.

````python
nombre = 'Pablo'

for i in range(len(nombre)):
    print(nombre[i])
````

````python
for letra in nombre:
    print(letra)
````

* Iteración de tuplas:

````python
tuplas = [(1, 2,4), (3, 4), (5, 6)]
for a, b,c in tuplas:
    print(a, b,c)
````

* Iteración de diccionarios. Se iteran las claves:

````python
diccionario = {'a': 1, 'b': 2, 'c': 3}
for key in diccionario:
    print(f"{key} => {diccionario[key]}")
````

* Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items
y values

````python
for key, value in diccionario.items():
    print(f"{key} => {value}")
````

````python
for value in diccionario.values():
    print(value)
````

* La variable item en la cabecera del for puede ser cualquier expresión que sea válida como
parte izquierda de una asignación convencional

````python
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
````

Las sentencias `break`, `continue` y `else`

Break y `continue`

- Solo tienen sentido dentro de bucles.
- `break` permite terminar el bucle por completo.
- `continue` permite saltar a la siguiente iteración, continuando con el bucle.
- Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (`if`).

````python
# Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]
for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
````

````python
for i in range(10):
    if i % 2 == 0:
        continue # Si el número es par, salta a la siguiente iteración␣
    print(f'Odd number {i}')
````

* Observa como la sentencia continue te puede ayudar a reducir el número de niveles de
anidamiento.
* Sin continue el anterior ejemplo sería:

````python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
````

````python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
````

### Else
* Los bucles pueden tener una sensencia else.
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros
lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de
un break.

````python
lista = [60,60,30,60]
element_to_find = 60
for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
````

````python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False
for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
````

List Comprehensions

- Permiten construir listas a través de la ejecución repetida (sentencia `for`) de una expressión para cada item de un objeto iterable.
- Van entre `[` y `]`. Esto es indicativo de que estamos construyendo una lista.

Sintaxis:

```python
[<expression> for <item> in <iterable>]
```

Ejemplo: repetir los carácteres de un string.

````python
[char*3 for char in "Enrique"]
````

````python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
````

* Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

````python
[2 for _ in 'Enrique']
````

Versión extendida

- Se puede especificar un filtro (sentencia `if`) para obtener únicamente los elementos que cumplan cierta condición.

Sintaxis:

```python
[<expression> for <item> in <iterable> if <condition>]
```

Ejemplo: obtener números pares.

````python
[y for y in range(9) if y % 2 == 0]
````

* Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle
equivalente.

````python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
````

Versión completa

- La sentencia `if` de una list comprehension también puede contener una expresión alternativa.

Sintaxis:

```python
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
```

Ejemplo: poner a cero los números pares.

````python
[0 if x % 2 == 0 else x for x in range(9)]
````

````python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
````

````python
lista = []
for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
````

````python
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
````

````python
lista = []
for x in range(-4,5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
````

Anidamiento

- Las list comprehensions soportan anidamiento en sus expresiones.

Ejemplo: bucle anidado para obtener una lista de tuplas que combinan los elementos de dos listas dadas.

````python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
````

### Pros y contras

- Principales ventajas de las comprehensions en comparación a un bucle convencional:
  - Expresión compacta y legible, si estás familiarizado con la sintaxis.
  - Mejor rendimiento.
- Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una expresión difícil de entender.

### Otros tipos de comprehensions

Dictionary comprehensions

- Usando `{` y `}` como delimitadores y una expresión `clave : valor`, se obtiene un diccionario en lugar de una lista.

Ejemplo: diccionario donde cada valor es el cuadrado de la clave.

````python
d = {x : x*x for x in range(10)}
print(d)
print(type(d))
````

````python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
shopping = {k:v for k,v in zip(items,price)}
print(shopping)
````

````python
items = ['Banana', 'Pear', 'Olives']
price = [1.1, 1.4, 2.4]
for k in zip(items, price):
    print(type(k))
    print(k)
````

Set comprehensions

- Usando `{` y `}` como delimitadores y una expresión simple (al igual que en las list comprehensions), se obtiene un conjunto.

Ejemplo: conjunto que incluye los 10 primeros números naturales.

````python
c = {x for x in range(10)}    # Crea un conjunto (set) con los valores del 0 al 9

print(c)                      # Muestra el contenido del conjunto

print(type(c))                # Muestra el tipo de dato de c
````

### Tuple comprehensions

````python
tupla = tuple(x for x in range(4))   # Crea una tupla con los valores generados desde 0 hasta 3

print(tupla)                         # Muestra el contenido de la tupla
````

#### Excepciones

- Las excepciones son eventos que representan situaciones excepcionales.
- Alteran el flujo de ejecución convencional.
- Python lanza excepciones automáticamente cuando se producen errores.
- El programador puede lanzar excepciones de manera explícita y también capturar excepciones para actuar como se crea conveniente.

Try/except

- Permite capturar excepciones y actuar en consecuencia.

Ejemplo: error de acceso fuera de rango.

````python
lista = [6, 1, 0, 5]        # Crea una lista con cuatro elementos

lista[4]                    # Intenta acceder al elemento con índice 4

print('Código tras el error') # Imprime el mensaje después del error
````

````python
lista = [6, 1, 0, 5]                          # Crea una lista con cuatro elementos

i = 300                                       # Define un índice que no existe en la lista

try:                                          # Inicia un bloque donde puede ocurrir un error
    a = lista[i]                              # Intenta acceder al elemento de la posición 300

except IndexError:                            # Captura el error de índice fuera de rango
    print('He capturado la excepción de tipo IndexError')  # Muestra mensaje de error controlado
    a = 0                                     # Asigna un valor a a para continuar el programa

print('Código tras el bloque try')            # Se ejecuta después del manejo del error

print(a)                                      # Muestra el valor final de a
````

* Normalmente, al capturar excepciones, querremos ser lo más específicos posible, pero también
se puedes usar una sentencia try-except que capture cualquier error.

````python
try:                                           # Inicia un bloque donde puede ocurrir un error
    4 / 0                                      # Intenta realizar una división entre cero

except ZeroDivisionError:                      # Captura el error producido por dividir entre cero
    print('He capturado la excepción de tipo ZeroDivisionError')  # Muestra un mensaje indicando que se controló el error

print('Código tras el bloque try')              # Se ejecuta después del manejo de la excepción
````

* Recoger la excepción e imprimir el mensaje de error

````python
lista = [6, 1, 0, 5]                         # Crea una lista con cuatro elementos (índices del 0 al 3)

try:                                         # Inicia un bloque donde puede ocurrir un error
    print(lista[4])                          # Intenta acceder al elemento de la posición 4 de la lista

except Exception as e:                       # Captura cualquier excepción y guarda el error en la variable e
    print("Error = " + str(e))               # Muestra el mensaje del error convertido a texto

print('Reachable Code')                      # Se ejecuta después del manejo del error
````

### Try/finally
* A través de finally podemos especificar código que queremos que se ejecute siempre (independientemente de si se produce la excepción o no).
* Se suele usar para liberar recursos.

````python
lista = [6, 1, 0, 5]                         # Crea una lista con cuatro elementos

try:                                         # Inicia un bloque donde puede ocurrir un error
    a = lista[1]                             # Intenta acceder al elemento de la posición 1 de la lista

except IndexError:                           # Captura el error si el índice está fuera del rango
    print('Exception IndexError Captured')   # Muestra un mensaje indicando que ocurrió un error

finally:                                     # Bloque que siempre se ejecuta, haya error o no
    print('Bloque finally')                  # Muestra un mensaje indicando que se ejecutó finally

b = lista[2]                                 # Accede al elemento de la posición 2 de la lista

print('Código tras el bloque try')            # Se ejecuta después del manejo de la excepción

print(b)                                      # Muestra el valor almacenado en b
````

### raise
* La sentencia raise nos permite lanzar excepciones de manera explícita.

````python
lista = [6, 1, 0, 5]                         # Crea una lista con cuatro elementos

indice = 4                                   # Define el índice que se quiere comprobar

try:                                         # Inicia un bloque donde puede ocurrir un error
    if indice >= len(lista):                 # Comprueba si el índice supera el tamaño de la lista
        raise IndexError('Índice fuera de rango')  # Genera manualmente una excepción de tipo IndexError

except IndexError:                           # Captura la excepción de tipo IndexError
    print('Excepción de tipo IndexError capturada')  # Muestra un mensaje indicando que se controló el error
````

````python
 # Ejercicio 7
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]
for i in a:
    if i > 5:
        for j in b:
            if j < 14:
                print(f"{i} x {j} = {i*j}")
````

````python
# Ejercicio 8
try:
    x = input()
    numero = float(x)
    print(f"El resultado de 10/{x} es: {10/numero}")
except ValueError:
    print(f"El valor '{x}' no es numérico")
except ZeroDivisionError:
    print("Error: división por cero")
````

````python
# Ejercicio 9
notas = {'Programación': 9.5, "Inteligencia artificial": 7.6, "Redes": 5.2,"Sistemas operativos": 6.4}
try:
    asignatura = input("Introduce una asignatura")
    print(f"La nota de la asignatura '{asignatura}' es: {notas[asignatura]}")
except KeyError:
    print(f"Asignatura incorrecta: {asignatura}")
````

````python
# Ejercicio 10
lista = [1, 4, -3, -1.5, 6.5, 2, 8, 2.1]
[numero for numero in lista if type(numero) == int and numero > 0]
````

````python
# Ejercicio 11
palabra = "murcielago"
{letra for letra in palabra if letra in ('a', 'e', 'i', 'o', 'u')}
````

````python
# Ejercicio 14
numero_a_palabra = {6 : "seis", 7 : "siete", 8 : "ocho", 9 : "nueve", 10 :"diez"}
[numero if numero <= 5 else numero_a_palabra[numero] for numero in range(1,11)]
````
