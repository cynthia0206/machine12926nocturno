## 1.1 ¿Qué son?

- Son un mecanismo que te ofrece el lenguaje para agrupar conjuntos de instrucciones de manera que se puedan ejecutar cuando el programador considere oportuno.

- Se debe usar este nombre para invocar (ejecutar) el código de la función.
- Las funciones calculan un resultado output en base a unos parámetros de entrada input.
- En cada invocación de la función, los parámetros de entrada pueden variar.

````python
import copy                                  # Importa el módulo copy para realizar copias.

lista = [2, 4, 16, 32, [34, 10, [5,5]]]      # Crea una lista con estructuras anidadas.

# copia = lista.copy()                       # Copia superficial (comentada).
copia = copy.deepcopy(lista)                 # Copia profunda de toda la estructura.

copia[0] = 454                               # Modifica el primer elemento de la copia.
copia[4][2][0] = 64                          # Modifica un elemento interno de la copia.

print(lista)                                 # Muestra la lista original.
print(copia)                                 # Muestra la lista modificada.

print(f"{id(lista)} - {id(copia)}")          # Muestra los IDs de memoria de ambas listas.
print(f"{id(lista[4])} - {id(copia[4])}")    # Muestra los IDs de los elementos internos.
````

## 1.6 Argumentos arbitrarios

- No se sabe a priori cuantos elementos se reciben.
- Se reciben como una tupla.
- Los parámetros se especifican con el símbolo `*`.

````python
def saludar(*names):                       # Define una función con argumentos arbitrarios.
    print(type(names))                     # Muestra el tipo de dato de names (tuple).
    for name in names:                     # Recorre cada nombre recibido.
        print("Hola " + name)              # Imprime un saludo para cada nombre.

saludar()                                  # Llama a la función sin argumentos.
saludar("Manolo", "Pepe", "Luis", "Alex", "Juan")  # Llama a la función con cinco nombres.
````

````python
def consulta_sql(**kwargs):                              # Define una función con argumentos arbitrarios por nombre.
    print(kwargs)                                        # Muestra el diccionario recibido.
    sql = "SELECT * FROM bicis WHERE "                   # Inicializa la consulta SQL.
    args = [f"{key}={value}" for key, value in kwargs.items()]  # Genera las condiciones de la consulta.
    sql += " AND ".join(args)                            # Une las condiciones con AND.
    return sql                                           # Devuelve la consulta SQL.

print(consulta_sql(station='alameda', free=10))          # Llama a la función e imprime la consulta generada.
````

### 1.7 Recursividad
* Funciones que se llaman a sí mismas

````python
def factorial(n):                    # Define función recursiva para calcular factorial
    if n == 0:                       # Caso base: si n es 0
        return 1                     # factorial de 0 es 1
    else:                            # Caso recursivo
        return n * factorial(n-1)    # multiplica n por factorial de n-1

print(factorial(10))                 # Imprime el factorial de 10


def fact(n):                         # Define función iterativa para factorial
    a = 1                            # Inicializa acumulador en 1
    for i in range(1, n+1):         # Recorre desde 1 hasta n
        a *= i                      # Multiplica acumulador por i
    return a                        # Devuelve el resultado final

print(fact(10))                     # Imprime el factorial de 10
````

````python
fact = lambda x: 1 if x == 0 else x * fact(x-1)   # Define una función lambda recursiva para el factorial

fact(10)                                           # Llama a la función y calcula el factorial de 10
````

## 1.8 Funciones Lambda

- Es posible definir funciones anónimas (sin nombre).
- Lambdas son expresiones; por lo tanto, puede aparecer en lugares donde una sentencia def no puede (por ejemplo, dentro de una lista o como parámetro de una función).
- Útiles cuando se quiere pasar una función como argumento a otra.

Formato general:

```python
lambda <lista de argumentos> : <valor a retornar>
```

- Importante: `<valor a retornar>` no es un conjunto de instrucciones. Es simplemente una expresión return que omite esta palabra.
- Las funciones definidas con def son más generales:
  - Cualquier cosa que implementes en un lambda, lo puedes implementar como una función convencional con def, pero no viceversa.

````python
# Ejemplo: función suma.
def suma(x, y, z):                 # Define una función que suma tres valores
    return x + y + z               # Devuelve la suma de los tres parámetros

print(suma(1, 2, 3))              # Llama a la función suma e imprime el resultado


# Ejemplo: función suma como expresión lambda.
suma2 = lambda x, y, z : x + y + z   # Define una función lambda que suma tres valores

print(suma2(1, 2, 3))              # Llama a la función lambda e imprime el resultado
````

````python
# La funcion 'sort' puede recibir una función optional como parámetro.
# Esta función 'key' se invoca para cada elemento de la lista antes de realizar las comparaciones.
# Ejemplo: ordenar strings por número de carácteres.

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']   # Lista de ciudades

cities.sort()                                          # Ordena alfabéticamente
print(cities)                                          # Imprime la lista ordenada

cities.sort(key = lambda x : len(x))                   # Ordena por número de caracteres usando lambda
print(cities)                                          # Imprime la lista ordenada

cities.sort(key = len)                                # Ordena por longitud usando la función len directamente
print(cities)                                          # Imprime la lista ordenada

def longitud(x):                                       # Define función que devuelve la longitud
    return len(x)

cities.sort(key = longitud)                           # Ordena usando una función definida por el usuario
print(cities)                                          # Imprime la lista ordenada
````

````python
# Mismo ejemplo sin lambda

cities = ['Valencia', 'Lugo', 'Barcelona', 'Madrid']   # Lista de ciudades

def num_caracteres(x):                                 # Define función que calcula longitud
    return len(x)                                      # Devuelve el número de caracteres

cities.sort(key = num_caracteres)                     # Ordena la lista usando la función definida
print(cities)                                         # Imprime la lista ordenada
````

* Diccionarios se pueden utilizar como ‘switch’ en otros lenguajes
* Ideal para escoger entre varias opciones

````python
def switch_dict(operator, x, y):                    # Define una función que recibe operador y dos valores
    # crea un diccionario de opciones y selecciona 'operator'
    return {
        'add': lambda: x + y,                      # Suma
        'sub': lambda: x - y,                      # Resta
        'mul': lambda: x * y,                      # Multiplicación
        'div': lambda: x / y,                      # División
    }.get(operator, lambda: None)()                 # Si no existe el operador devuelve None

print(switch_dict('add', 2, 2))                    # Llama suma
print(switch_dict('div', 10, 2))                   # Llama división
print(switch_dict('mod', 17, 3))                   # Operador no existe → devuelve None
````

## 1.9 Scopes

- La localización de una asignación a una variable en el código determina desde dónde puedes acceder a esa variable.
- Por ejemplo, una variable asignada dentro de una función sólo es visible dentro de esa función.
- El alcance de la visibilidad de una variable determina su *scope*.
- El término *scope* hace referencia a un espacio de nombres (*namespace*). Por ejemplo, una función establece su propio *namespace*.

````python
X = 10                      # Variable global con valor 10

def func():                # Define función externa
    X = 20                 # Variable local de func (no modifica la global)
    
    def func_int():        # Define función interna (anidada)
        X = 30             # Variable local de func_int
        print(X)           # Imprime la X local (30)
    
    func_int()            # Llama a la función interna

func()                    # Llama a la función principal
````

## Regla LEGB

Dada una función F, tenemos los siguientes scopes:

- Local: variables locales a F.
- Enclosing: variables localizadas en funciones que contienen a (o por encima de) F.
- Global: variables definidas en el módulo (fichero) que no están contenidas en ninguna función. Este scope no abarca más de un fichero.
- Built-ins: proporcionadas por el lenguaje.

La resolución de nombres ocurre de abajo a arriba.

````python
# Global (X y func)
X = 99                              # Variable global

def func(Y):                        # Función con parámetro local Y
    Z = X + Y                       # Usa variable global X y parámetro Y
    return Z                        # Devuelve el resultado

print(func(1))                      # Llama a la función e imprime el resultado
````

````python
X = 1                  # Variable global X con valor 1

def func():           # Definición de la función
    X = 2             # Variable local X (diferente a la global)

func()                # Se llama a la función (no modifica la global)

print(X)              # Imprime la variable global X
````

* La sentencia global nos permite modificar una variable global desde dentro de una función

````python
X = 1              # Variable global inicializada con valor 1

def func():        # Definición de la función
    global X       # Indica que se usará la variable global X
    X = 2          # Modifica la variable global X

func()             # Llama a la función (cambia X global)

print(X)           # Imprime la variable global modificada
````

* No hay necesidad de usar global para referenciar variables. Sólo es necesario para modificarlas.

````python
y, z = 1, 2                 # Variables globales y y z

def todas_globales():       # Función principal
    global x                # Declara x como variable global
    x = y + z              # Suma variables globales y asigna a x

todas_globales()            # Ejecuta la función

print(x)                    # Imprime x (global)
print(y)                    # Imprime y (global)
print(z)                    # Imprime z (global)

X = 77                      # Variable global

def f1():                   # Función externa
    X = 88                  # Variable local (enclosing scope para f2)

    def f2():              # Función interna
        print(X)           # Accede a la variable del enclosing scope

    f2()                   # Llama a la función interna

f1()                        # Llama a la función principal
````

## Closures

- Las funciones pueden recordar su enclosing scope, independientemente de si estos continúan existiendo o no.

````python
def f1():                     # Define función externa
    X = 88                   # Variable local (enclosing scope para f2)

    def f2():                # Función interna
        print(X)             # Accede a la variable del enclosing scope

    return f2                # Devuelve la función sin ejecutarla

accion = f1()                # f1 devuelve la función f2 (closure)
accion()                     # Se ejecuta la función devuelta
````

* Sentencia nonlocal para modificar variables que no son locales, pero tampoco globales.

````python
def f1():                 # Función externa
    contador = 10        # Variable del enclosing scope

    def f2():            # Función interna (closure)
        nonlocal contador  # Permite modificar la variable del enclosing scope
        contador += 1     # Incrementa el contador
        print(contador)  # Imprime el valor actualizado

    return f2            # Devuelve la función sin ejecutarla

accion = f1()            # Se obtiene la función f2 con su estado

accion()                 # 11
accion()                 # 12
accion()                 # 13
````

````python
# Ejercicio 1
# Función que recibe una lista y devuelve los cuadrados

def cuadrados(lista):
    return [x ** 2 for x in lista]

print("Ejercicio 1:", cuadrados([1, 2, 3, 4]))
````

````python
# Ejercicio 2
# Función que suma una cantidad indeterminada de números

def suma(*numeros):
    return sum(numeros)

print("Ejercicio 2:", suma(1, 2, 3, 4, 5))
````

````python
# Ejercicio 3
# Función que invierte un string

def invertir(cadena):
    return cadena[::-1]

print("Ejercicio 3:", invertir("hola"))
````

````python
# Ejercicio 4
# Función lambda que invierte un string

invertir_lambda = lambda s: s[::-1]

print("Ejercicio 4:", invertir_lambda("hola"))
````

````python
# Ejercicio 5
# Función que comprueba si un número está en un rango

def en_rango(n, minimo, maximo):
    return minimo <= n <= maximo

print("Ejercicio 5:", en_rango(5, 1, 10))
````

````python
# Ejercicio 6
# Función que devuelve los 5 primeros múltiplos de un número positivo

def multiplos(n):
    if n <= 0:
        print("Error: número inválido")
        return []
    return [n * i for i in range(1, 6)]

print("Ejercicio 6:", multiplos(3))
````

````python
# Ejercicio 7
# Función que detecta duplicados en una lista

def tiene_duplicados(lista):
    return len(lista) != len(set(lista))

print("Ejercicio 7:", tiene_duplicados([1, 2, 3, 3]))
````

````python
# Ejercicio 8
# Lambda que detecta duplicados en una lista

duplicados_lambda = lambda lista: len(lista) != len(set(lista))

print("Ejercicio 8:", duplicados_lambda([1, 2, 3, 4]))
````

````python
# Ejercicio 9
# Función que comprueba si un string es palíndromo

def palindromo(texto):
    texto = texto.lower()
    return texto == texto[::-1]

print("Ejercicio 9:", palindromo("reconocer"))
print("Ejercicio 9:", palindromo("python"))
````
