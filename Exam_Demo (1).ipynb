```python
# Your solution HERE !!!!

def extract_year(movies):                  # Define una función llamada extract_year que recibe 'movies'

    def split_year(title):                 # Función interna para extraer el año del título
        index = title.index('-')           # Busca la posición donde aparece el carácter "-"
        return title[index:].replace('-','').strip()  
                                            # Toma el texto desde "-" hasta el final,
                                            # elimina el "-" y quita espacios en blanco.
                                            # Ejemplo: "Toy Story - 1995" → "1995"

    def remove_year(title):                # Función interna para eliminar el año del título
        index = title.index('-')           # Busca la posición del carácter "-"
        return title[:index-1].strip()     # Devuelve el texto antes del "-"
                                            # (restando 1 para quitar el espacio anterior)
                                            # y elimina espacios sobrantes.
                                            # Ejemplo: "Toy Story - 1995" → "Toy Story"

    df_movies = pd.DataFrame(movies)       # Convierte 'movies' en un DataFrame de pandas

    df_movies['Year'] = df_movies['Name'].apply(split_year)
                                            # Aplica la función split_year a cada valor
                                            # de la columna 'Name' y guarda el resultado
                                            # en una nueva columna llamada 'Year'

    df_movies['Name'] = df_movies['Name'].apply(remove_year)
                                            # Aplica la función remove_year a cada valor
                                            # de la columna 'Name' para dejar solo
                                            # el nombre de la película

    return df_movies                       # Devuelve el DataFrame con las columnas
                                            # 'Name' (sin año) y 'Year'
```

```python
def extract_year(movies):                     # Define una función llamada extract_year que recibe un diccionario o DataFrame llamado movies.

    nombres = []                              # Crea una lista vacía para guardar los nombres de las películas.

    anios = []                                # Crea una lista vacía para guardar los años de las películas.

    for titulo in movies['Name']:             # Recorre uno por uno los títulos que están en la columna o clave 'Name'.

        partes = titulo.rsplit(' - ', 1)      # Divide el texto desde la derecha usando " - " como separador.
                                              # El 1 indica que solo hará una división.
                                              # Ejemplo:
                                              # "Spider-Man - No Way Home - 2021"
                                              # Resultado:
                                              # ["Spider-Man - No Way Home", "2021"]

        nombre = partes[0]                    # Guarda la primera parte (el nombre de la película).

        anio = partes[1]                      # Guarda la segunda parte (el año).

        nombres.append(nombre)                # Agrega el nombre de la película a la lista nombres.

        anios.append(anio)                    # Agrega el año a la lista anios.

    return pd.DataFrame({'Name': nombres, 'Year': anios})
                                              # Crea un DataFrame con las listas nombres y anios
                                              # y lo devuelve como resultado.
```

```python
def extract_year(movies):                                  # Define una función llamada extract_year que recibe los datos de las películas.

    df_movies = pd.DataFrame(movies)                       # Convierte 'movies' en un DataFrame de pandas.

    df_movies['Year'] = df_movies['Name'].apply(           # Crea una nueva columna llamada 'Year'.
        lambda x: x.split('-')[1].strip()                  # Para cada título (x):
                                                           # 1. split('-') divide el texto por el guion.
                                                           # 2. [1] toma la segunda parte (el año).
                                                           # 3. strip() elimina los espacios.
                                                           # Ejemplo:
                                                           # "Toy Story - 1995"
                                                           # split('-') → ['Toy Story ', ' 1995']
                                                           # [1] → ' 1995'
                                                           # strip() → '1995'
    )

    df_movies['Name'] = df_movies['Name'].apply(           # Modifica la columna 'Name'.
        lambda x: x.split('-')[0].strip()                  # Para cada título (x):
                                                           # 1. split('-') divide el texto por el guion.
                                                           # 2. [0] toma la primera parte (el nombre).
                                                           # 3. strip() elimina los espacios.
                                                           # Ejemplo:
                                                           # "Toy Story - 1995"
                                                           # split('-') → ['Toy Story ', ' 1995']
                                                           # [0] → 'Toy Story '
                                                           # strip() → 'Toy Story'
    )

    return df_movies                                       # Devuelve el DataFrame con las columnas Name y Year separadas.
```

```python
def find_duplicates(numbers):
    _, idx = np.unique(numbers, return_index = True)
    result = np.zeros_like(numbers, dtype = bool)
    result[idx] = True
    result = np.invert(result)
    return result
```

```python
def find_duplicates(numbers):
  visto = set()
  output = []
  for value in numbers:
    if value in visto:
      output.append(True)
    else:
      visto.add(value)
      output.append(False)
  return np.array(output)
```

## 03
- Resolución de dos sistemas de ecuaciones
    - $ A : \begin{cases} -9x +  4y = 20 \\ -7y + 16x = 80                         \end{cases}$
    
    - $ B : \begin{cases} x - 2y + 3z = 7 \\ 2x + y + z = 4 \\ -3x + 2y -2z = -10 \end{cases}$
- Verifica por código los resultados

```python
# Your solution HERE !!!!                   

import math                                   # Importa la librería math para utilizar funciones matemáticas.

import numpy.linalg as linalg                 # Importa el módulo de álgebra lineal de NumPy y le asigna el alias linalg.

a = np.array([[-9, 4], [16, -7]])             # Crea una matriz de coeficientes de 2x2 para representar un sistema de ecuaciones.

b = np.array([20, 80])                        # Crea un vector con los resultados (lado derecho) de las ecuaciones.

print(a)                                      # Muestra la matriz de coeficientes.

print(b)                                      # Muestra el vector de resultados.

x, y = linalg.solve(a, b)                     # Resuelve el sistema de ecuaciones y guarda la solución en x e y.

print(f'x={x} y={y}')                         # Imprime los valores encontrados para x e y.

print(math.isclose(-9*x + 4*y, 20))           # Sustituye x e y en la primera ecuación y verifica si el resultado es 20.
                                               # Devuelve True si la solución es correcta.

print(math.isclose(16*x - 7*y, 80))           # Sustituye x e y en la segunda ecuación y verifica si el resultado es 80.
                                               # Devuelve True si la solución es correcta.

A2 = np.array([[1, -2, 3],                    # Crea una matriz de coeficientes de 3x3 para un sistema de tres ecuaciones.
               [2, 1, 1],
               [-3, 2, -2]])

b2 = np.array([7, 4, -10])                    # Crea el vector con los resultados de las tres ecuaciones.

x, y, z = np.linalg.solve(A2, b2)             # Resuelve el sistema de tres ecuaciones y guarda los resultados en x, y y z.

print(f'x={x} y={y} z={z}')                   # Muestra los valores encontrados para x, y y z.

print(math.isclose(1*x - 2*y + 3*z, 7))       # Comprueba que la primera ecuación se cumple con los valores obtenidos.
                                               # Si es correcta imprime True.

print(math.isclose(2*x + 1*y + z, 4))         # Comprueba que la segunda ecuación se cumple.
                                               # Si es correcta imprime True.

print(math.isclose(-3*x + 2*y - 2*z, -10))    # Comprueba que la tercera ecuación se cumple.
                                               # Si es correcta imprime True.
```

- Escribir un generador `frange` que simule el comportamiento de la función `arange` de NumPy
- Esta función cada vez que se invoque debe devolver valores espaciados uniformemente a partir de un valor inicial. Se debe proveer a dicha función un valor inicial para la secuencia de números y un valor para el step
- Debe soportar valores númericos en coma flotante
- La función potencialmente se puede llamar un número infinito de veces
- Usando plain Python, sin ninguna librería

```python
import math

init = 10.6
step = 0.8
g = frange(init, step)  # This calls your function frange

iters = 10
numbers = [next(g) for n in range(iters)]
for i, n in enumerate(numbers):
    assert(math.isclose(n, init + (step * i)))  # This will fail if the range is not properly created
print(numbers)
```

```python
# Your solution HERE !!!!                    

def frange(init, step):                       # Define una función generadora llamada frange.
                                              # Recibe dos parámetros:
                                              # init: valor inicial de la secuencia.
                                              # step: incremento que se sumará en cada paso.

    while True:                               # Inicia un ciclo infinito. Se repetirá hasta que
                                              # el usuario deje de solicitar valores.

        yield init                            # Devuelve el valor actual de init y pausa la función.
                                              # Cuando se vuelva a llamar al generador,
                                              # continuará desde la siguiente línea.

        init += step                          # Suma el valor de step a init para generar
                                              # el siguiente número de la secuencia.
```
